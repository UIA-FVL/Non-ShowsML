```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('database_non-shows.xlsx')
```

Age distribution by last appointment status

```python
df['Appointment Type'] = df['Appointment Type'].replace({1: 'Attendance', 0: 'Absence'})
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.violinplot(x='Appointment Type', y='Age', data=df, hue='Appointment Type', palette="muted")
plt.title('Age distribution by last appointment status')
plt.xlabel('Last appointment status')
plt.ylabel('Age')
plt.show()
```

Gender distribution of patients

```python
conteo_sexo = df['Sex'].value_counts()
sns.set(style="whitegrid")
colores = sns.color_palette('pastel')
nuevas_etiquetas = ['Female', 'Male', 'Undetermined']
plt.figure(figsize=(8, 8))
plt.pie(conteo_sexo, labels=nuevas_etiquetas, autopct='%1.1f%%', startangle=140, colors=colores, wedgeprops={'linewidth': 0.5, 'edgecolor': 'black'})
plt.title('Gender distribution of patients')
plt.axis('equal')
plt.show()
```

Gender distribution by type of appointment

```python
last_appointment_etiquetas = {0: 'Absence', 1: 'Attendance'}
df['Last appointment status'] = df['Last appointment status'].map(last_appointment_etiquetas)
sexo_etiquetas = {0: 'Female', 1: 'Male', 2: 'Undetermined'}
df['Gender'] = df['Gender'].map(sexo_etiquetas)
sns.set(style="whitegrid")
colores = sns.color_palette('pastel')
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Last appointment status', hue='Gender', palette=colores)
plt.title('Gender distribution by type of appointment')
plt.xlabel('Last appointment status')
plt.ylabel('Count')
plt.show()
```

Distribution of type of insurance company by last appointment status

```python
last_appointment_etiquetas = {0: 'Absence', 1: 'Attendance'}
df['Last appointment status'] = df['Last appointment status'].map(last_appointment_etiquetas)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Insurance Type', hue='Last appointment status', palette='Set2')
plt.title('Distribution of type of insurance company by last appointment status')
plt.xlabel('Insurance type')
plt.ylabel('Count')
plt.show()
```

Distribution of the time interval between creation and assignment of non-attendance appointments

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_tipo_cita_0 = df[df['Appointment Type'] == 0]
plt.figure(figsize=(10, 6))
sns.histplot(data=df_tipo_cita_0, x='Creation to Assignment Interval', bins=10, color='blue', edgecolor='black')
plt.title('Distribution of the time interval between creation and assignment of non-attendance appointments')
plt.xlabel('Day interval')
plt.ylabel('Frecuency')
plt.grid(True)
plt.show()
```

Attendance and Absences Values by Data Balanced Method

```python
metodos = ['Unbalanced', 'RandomSplit', 'SMOTE', 'ADASYN']
attended = [12153, 6434, 9697, 9697]
absences = [6434, 6434, 9697, 9818]

# Configuracin del grfico
plt.figure(figsize=(10, 6))

# Ancho de las barras
bar_width = 0.35

# Posicin de las barras
index = np.arange(len(metodos))

# Graficar barras de asistencias
plt.bar(index, attended, bar_width, color='skyblue', label='Attended')

# Graficar barras de ausencias
plt.bar(index + bar_width, absences, bar_width, color='salmon', label='Absences')

# Etiquetas de valores numricos
for i in range(len(metodos)):
    plt.text(i, attended[i] + 100, str(attended[i]), ha='center', fontsize=10)
    plt.text(i + bar_width, absences[i] + 100, str(absences[i]), ha='center', fontsize=10)

# Ttulos y etiquetas
plt.title('Attendance and Absences Values')
plt.xlabel('Data Balanced Method')
plt.xticks(index + bar_width / 2, metodos)
plt.legend()

# Ocultar la escala numrica del eje y
plt.yticks([])

# Mostrar grfico
plt.show()
```

Status of the last appointment

```python
# Contar la frecuencia de cada clase en 'Tipo de cita'
frecuencia_clases = df['Appointment Type'].value_counts()
# Calcular el porcentaje de cada clase
porcentaje_clases = frecuencia_clases / len(df) * 100
# Cambiar el nombre de las clases
nombres_clases = {0: 'Absences', 1: 'Attendances'}
porcentaje_clases.index = porcentaje_clases.index.map(nombres_clases)


# Definir colores un poco ms oscuros para cada clase
colores_clases = ['#66ff66','#66b3ff']

# Crear el diagrama de torta
plt.figure(figsize=(8, 6))
plt.pie(porcentaje_clases, labels=porcentaje_clases.index, autopct='%1.1f%%', startangle=140, colors=colores_clases)
plt.title('Status of the last appointment')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
