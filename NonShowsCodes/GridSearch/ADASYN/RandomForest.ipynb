```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import ADASYN
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

ADASYN_NonSD

```python
data = pd.read_excel('database_non-shows.xlsx')
```

```python
x = data.drop('Appointment Type', axis=1)
y = data['Appointment Type']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
adasyn = ADASYN(sampling_strategy='minority', random_state=45)
X_train, y_train = adasyn.fit_resample(X_train, y_train)
```

```python
param_grid = {
    'n_estimators' : [10, 50, 100, 200],
    'criterion' : ["gini", "entropy"],
    'max_depth' : [10, 20, 30],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4],
    'max_features' : [None, "auto", "sqrt", "log2", 2, 3]
}

model = RandomForestClassifier()
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_accuracy = grid_search.best_score_

best_model = RandomForestClassifier(**best_params)
best_model.fit(X_train, y_train)


print("Mejor combinacin de hiperparmetros:", best_params)
print("Precisin ms alta encontrada:", best_accuracy)
```
